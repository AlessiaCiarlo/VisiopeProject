```python
data_dir_path = '/content/drive/MyDrive/VISIOPE_Project/DATASET/CLASSIFICATION_dataset/'

dict_label = {'biodegradable': 0, 'glass' : 1, 'paper' : 2, 'plastic' : 3, 'special' : 4, 'trash' : 5}

my_list = os.listdir(data_dir_path)

train = []
val = []
test = []

for dir in my_list:
  type_list = os.listdir(data_dir_path+dir)
  print(dir)
  random.shuffle(type_list)

  # Calculate the size of each sublist

  train_size = int(0.7 * len(type_list))
  val_size = int(0.15 * len(type_list))
  test_size = len(type_list) - train_size - val_size

  # Split the list into three sublists
  sublist1 = type_list[:train_size]
  sublist2 = type_list[train_size:train_size+val_size]
  sublist3 = type_list[train_size+val_size:]

  list_ = [sublist1, sublist2, sublist3]

  for l in list_:

    print(len(l))

    for i in range(len(l)):
      name = l[i]
      l.pop(i)
      l.insert(i, dir+'/'+name)

    print(l)
    print(len(l))

  print('train', sublist1)
  print('val', sublist2)
  print('test', sublist3)
  for elem in sublist1: train.append(elem)
  for elem in sublist2: val.append(elem)
  for elem in sublist3: test.append(elem)
```

```python
print('train', train)
print(len(train))
print('val', val)
print(len(val))
print('test', test)
print(len(test))
```

```python
# Specify the file name
root_dir = '/content/drive/MyDrive/VISIOPE_Project/DATASET/CLASSIFICATION_dataset/'
csv_file_name = [('train_class.csv', train), ('val_class.csv', val), ('test_class.csv', test)]

for dir, l in csv_file_name:
  # Open the CSV file in write mode
  with open(root_dir+dir, 'w', newline='') as csvfile:
      # Create a CSV writer object
      csv_writer = csv.writer(csvfile)
      for row in l:
        # Write the data to the CSV file
        #print(row, dict_label[row.split('/')[0]])
        csv_writer.writerow([row, dict_label[row.split('/')[0]]])

  print(f'CSV file "{csv_file_name}" has been created.')
```

```python
# Specify the file name
root_dir = '/content/drive/MyDrive/VISIOPE_Project/DATASET/CLASSIFICATION_dataset/'
csv_file_name = [('train_class', 1450), ('val_class', 300), ('test_class', 300)]

for dir, l in csv_file_name:
  # Open the CSV file in write mode
  with open(root_dir+dir+'.csv', 'r', newline='') as csvfile:
      # Create a CSV writer object
      csv_reader = csv.reader(csvfile)

      plastic = 0
      trash = 0
      with open(root_dir+dir+'_balanced.csv', 'w', newline='') as newcsv:
          csv_writer = csv.writer(newcsv)
          for row in csv_reader:
              if row[1] == '3':
                if plastic < l:
                  csv_writer.writerow(row)
                  plastic += 1
              elif row[1] == '5':
                if trash < l:
                  csv_writer.writerow(row)
                  trash += 1
              else: csv_writer.writerow(row)

  print(f'CSV file "{csv_file_name}" has been created.')
```

```python
filename = '/content/drive/MyDrive/VISIOPE_Project/DATASET/TACO_MJU/annotations.json'

new_filename = '/content/drive/MyDrive/VISIOPE_Project/DATASET/TACO_MJU/annotations_resized_final.json'

list_ = []

with open(filename, "r") as json_file:
  # Load and parse the JSON data
  data = json.load(json_file)

  c = 0
  for line in data:
    id = c
    filename = line['filename']
    width = line['width']
    height = line['height']
    seg = line['seg']
    area = line['area']
    bbox = line['bbox']
    iscrowd = 0

    new_width = 640
    new_height = 480
    new_seg = []
    for i in range(len(seg)):
      if i%2 == 0:   #x
        elem = seg[i]*new_width/width
      else:  #y
        elem = seg[i]*new_height/height
      new_seg.append(elem)
    print(len(seg), len(new_seg))
    new_area = area*(new_width*new_height)/(width*height)
    new_bbox = [bbox[0]*new_width/width, bbox[1]*new_height/height, bbox[2]*new_width/width, bbox[3]*new_height/height]

    list_.append({'id': id, 'filename':filename, 'width':new_width, 'height':new_height, 'seg':new_seg, 'area':new_area, 'bbox':new_bbox, 'iscrowd':iscrowd})
    c += 1

  with open(new_filename, "w") as new_file:
    json.dump(list_, new_file, indent=2)
```
